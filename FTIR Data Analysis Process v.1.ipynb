```python
from pathlib import Path
import re
from collections import defaultdict

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline

from brukeropus import read_opus


# ============================================================
# USER SETTINGS
# ============================================================
data_folder = Path(r"H:\Extra\Extra of 01.Apr")

sample_prefix = "BlueBronzeBB1Longv2"
background_prefix = "GoldMirror"

spectrum_block_name = "sm"

#   "ratio"         -> sample / background
#   "subtract"      -> sample - background
#   "no_background" -> use sample only, interpolate to common axis, then average
correction_mode = "ratio"

# Plotting windows
# Format: (window_name, x_min, x_max)
x_windows = [
    ("full", 250, 12000),
    ("low", 420, 1440),
    ("mid", 3000, 5000),
    ("high", 5000, 8000),
]

plt.rcParams["figure.figsize"] = (9, 6)
plt.rcParams["axes.grid"] = True

eps = 1e-30

# ============================================================
# OPTIONAL SAVING
# ============================================================
save_outputs = True

# Output folder for csv + figures
output_folder = data_folder / "processed_outputs"

# Save one averaged spectrum file per polarization
save_avg_spectra_per_pol = True

# Save the plot with all averaged polarization curves
save_main_average_plot = True

# Save the optional single-polarization plot
save_single_pol_plot = True

# Default dpi for saved figures
figure_dpi = 300


# ============================================================
# HELPERS
# ============================================================
def extract_measurement_dict_from_history(history_text):
    if not isinstance(history_text, str) or len(history_text.strip()) == 0:
        return {}

    match = re.search(r"MeasureSample\s*\(.*?\{(.*)\}\);", history_text, re.DOTALL)
    if not match:
        return {}

    param_block = match.group(1)
    pairs = re.findall(r"(\w+)\s*=\s*('.*?'|[^,]+)", param_block)

    out = {}
    for key, value in pairs:
        value = value.strip()
        if value.startswith("'") and value.endswith("'"):
            value = value[1:-1]
        else:
            try:
                if "." in value:
                    value = float(value)
                else:
                    value = int(value)
            except Exception:
                pass
        out[key] = value
    return out


def parse_filename_info(filename):
    """
    Examples:
      BlueBronzeBB1-Pol20deg-Run3.0
      GoldMirror-Pol20deg-Run3.0
    """
    stem = filename

    m = re.match(r"^(.*?)-Pol([\-+]?\d+(?:\.\d+)?)deg-Run(\d+)\.0$", stem, re.IGNORECASE)
    if not m:
        raise ValueError(f"Filename does not match expected pattern: {filename}")

    base_name = m.group(1)
    pol_deg = float(m.group(2))
    run_idx = int(m.group(3))

    return base_name, pol_deg, run_idx


def read_one_opus_file(filepath, spectrum_block="sm"):
    opus_data = read_opus(str(filepath))

    if not hasattr(opus_data, spectrum_block):
        raise ValueError(f"{filepath.name}: OPUS file does not contain block '{spectrum_block}'")

    block = getattr(opus_data, spectrum_block)
    x = np.asarray(block.x, dtype=float)
    y = np.asarray(block.y, dtype=float)

    idx = np.argsort(x)
    x = x[idx]
    y = y[idx]

    history_text = getattr(opus_data, "history", "")
    meta = extract_measurement_dict_from_history(history_text)

    base_name, pol_deg, run_idx = parse_filename_info(filepath.name)

    row = {
        "filename": filepath.name,
        "filepath": str(filepath),
        "base_name": base_name,
        "pol_deg": pol_deg,
        "run_idx": run_idx,
        "OPUS_POL": meta.get("POL", None),
        "NSS": meta.get("NSS", None),
        "NSR": meta.get("NSR", None),
        "RES": meta.get("RES", None),
        "LFQ": meta.get("LFQ", None),
        "HFQ": meta.get("HFQ", None),
        "LQ2": meta.get("LQ2", None),
        "HQ2": meta.get("HQ2", None),
        "SNM": meta.get("SNM", None),
        "EXP": meta.get("EXP", None),
        "DTC": meta.get("DTC", None),
        "SRC": meta.get("SRC", None),
        "VEL": meta.get("VEL", None),
    }

    return {
        "file": filepath,
        "x": x,
        "y": y,
        "meta": meta,
        "row": row,
        "opus_obj": opus_data,
    }


def load_all_runs(folder, spectrum_block="sm"):
    files = sorted(folder.glob("*.0"))
    if len(files) == 0:
        raise FileNotFoundError(f"No .0 files found in {folder}")

    runs = []
    rows = []

    for fp in files:
        run = read_one_opus_file(fp, spectrum_block=spectrum_block)
        runs.append(run)
        rows.append(run["row"])

    df = pd.DataFrame(rows)
    return runs, df


def build_global_common_axis(runs):
    """
    Build a single common axis for all runs based on:
      dx = min(RES)/2
      overlap region across all runs
    """
    xmins = [np.min(r["x"]) for r in runs]
    xmaxs = [np.max(r["x"]) for r in runs]

    x_low = max(xmins)
    x_high = min(xmaxs)

    if not (x_high > x_low):
        raise ValueError("No common overlap across all runs.")

    resolutions = []
    for r in runs:
        res = r["row"].get("RES", None)
        if res is None or not np.isfinite(res):
            raise ValueError(f"Missing RES in metadata for file {r['file'].name}")
        resolutions.append(float(res))

    dx = min(resolutions) / 2.0

    if dx >= 1:
        decimals = 0
    else:
        decimals = int(np.ceil(-np.log10(dx))) + 1

    start = np.ceil(x_low / dx) * dx
    stop = np.floor(x_high / dx) * dx

    x_common = np.arange(start, stop + 0.5 * dx, dx)
    x_common = np.round(x_common, decimals=decimals)

    return x_common, dx, decimals


def interp_to_axis(x, y, x_common):
    return np.interp(x_common, x, y)


def split_sample_background(runs, sample_prefix, background_prefix):
    samples = []
    backgrounds = []

    for r in runs:
        base = r["row"]["base_name"]
        if base == sample_prefix:
            samples.append(r)
        elif base == background_prefix:
            backgrounds.append(r)

    if len(samples) == 0:
        raise ValueError(f"No sample files found with prefix '{sample_prefix}'")
    if len(backgrounds) == 0:
        raise ValueError(f"No background files found with prefix '{background_prefix}'")

    return samples, backgrounds


def build_run_lookup(runs):
    """
    key = (pol_deg, run_idx)
    """
    lookup = {}
    for r in runs:
        key = (r["row"]["pol_deg"], r["row"]["run_idx"])
        if key in lookup:
            raise ValueError(f"Duplicate run key found: {key}")
        lookup[key] = r
    return lookup


def pair_sample_background_runs(sample_runs, background_runs):
    sample_lookup = build_run_lookup(sample_runs)
    background_lookup = build_run_lookup(background_runs)

    common_keys = sorted(set(sample_lookup.keys()) & set(background_lookup.keys()))
    if len(common_keys) == 0:
        raise ValueError("No matching sample/background pairs found by (pol_deg, run_idx).")

    missing_bg = sorted(set(sample_lookup.keys()) - set(background_lookup.keys()))
    missing_sample = sorted(set(background_lookup.keys()) - set(sample_lookup.keys()))

    if len(missing_bg) > 0:
        print("Warning: missing background for sample runs:", missing_bg)
    if len(missing_sample) > 0:
        print("Warning: missing sample for background runs:", missing_sample)

    pairs = []
    for key in common_keys:
        pairs.append({
            "key": key,
            "sample": sample_lookup[key],
            "background": background_lookup[key],
        })

    return pairs


def correct_one_pair(pair, x_common, mode="ratio"):
    s = pair["sample"]
    b = pair["background"]

    y_s = interp_to_axis(s["x"], s["y"], x_common)
    y_b = interp_to_axis(b["x"], b["y"], x_common)

    if mode == "ratio":
        y_corr = y_s / np.maximum(y_b, eps)
    elif mode == "subtract":
        y_corr = y_s - y_b
    else:
        raise ValueError("mode must be 'ratio' or 'subtract'")

    return {
        "pol_deg": s["row"]["pol_deg"],
        "run_idx": s["row"]["run_idx"],
        "x": x_common,
        "y_sample": y_s,
        "y_background": y_b,
        "y_corrected": y_corr,
        "sample_file": s["file"].name,
        "background_file": b["file"].name,
        "sample_meta": s["meta"],
        "background_meta": b["meta"],
        "sample_row": s["row"],
        "background_row": b["row"],
    }


def process_one_sample_run_no_background(sample_run, x_common):
    y_s = interp_to_axis(sample_run["x"], sample_run["y"], x_common)

    return {
        "pol_deg": sample_run["row"]["pol_deg"],
        "run_idx": sample_run["row"]["run_idx"],
        "x": x_common,
        "y_sample": y_s,
        "y_background": None,
        "y_corrected": y_s,
        "sample_file": sample_run["file"].name,
        "background_file": None,
        "sample_meta": sample_run["meta"],
        "background_meta": None,
        "sample_row": sample_run["row"],
        "background_row": None,
    }


def group_corrected_runs_by_polarization(corrected_runs):
    grouped = defaultdict(list)
    for r in corrected_runs:
        grouped[r["pol_deg"]].append(r)
    return dict(sorted(grouped.items(), key=lambda kv: kv[0]))


def average_corrected_runs(run_group):
    ys = np.asarray([r["y_corrected"] for r in run_group], dtype=float)
    x = run_group[0]["x"]
    n_runs = ys.shape[0]

    y_mean = np.mean(ys, axis=0)

    if n_runs > 1:
        y_std = np.std(ys, axis=0, ddof=1)
        y_sem = y_std / np.sqrt(n_runs)
    else:
        y_std = np.full_like(y_mean, np.nan)
        y_sem = np.full_like(y_mean, np.nan)

    return {
        "x": x,
        "y_all": ys,
        "y_mean": y_mean,
        "y_std": y_std,
        "y_sem": y_sem,
        "n_runs": n_runs,
        "run_indices": [r["run_idx"] for r in run_group],
        "sample_files": [r["sample_file"] for r in run_group],
        "background_files": [r["background_file"] for r in run_group],
    }


def pol_to_label(pol):
    if float(pol).is_integer():
        return f"{int(pol)}"
    return str(pol).replace(".", "p")


def ensure_output_dirs(base_folder):
    base_folder.mkdir(parents=True, exist_ok=True)
    spectra_dir = base_folder / "averaged_spectra"
    figures_dir = base_folder / "figures"
    spectra_dir.mkdir(parents=True, exist_ok=True)
    figures_dir.mkdir(parents=True, exist_ok=True)
    return spectra_dir, figures_dir


def save_avg_spectra(avg_dict, spectra_dir, correction_mode):
    for pol, d in avg_dict.items():
        df_out = pd.DataFrame({
            "wavenumber_cm^-1": d["x"],
            "y_mean": d["y_mean"],
            "y_std": d["y_std"],
            "y_sem": d["y_sem"],
        })
        df_out["n_runs"] = d["n_runs"]
        df_out["pol_deg"] = pol
        df_out["correction_mode"] = correction_mode

        outfile = spectra_dir / f"avg_spectrum_pol_{pol_to_label(pol)}deg_{correction_mode}.csv"
        df_out.to_csv(outfile, index=False)

    print(f"Saved averaged spectra per polarization to: {spectra_dir}")


def get_window_mask(x, x_min, x_max):
    mask = np.ones_like(x, dtype=bool)
    if x_min is not None:
        mask &= (x >= x_min)
    if x_max is not None:
        mask &= (x <= x_max)
    return mask


def make_window_label(window_name, x_min, x_max):
    x1 = "None" if x_min is None else str(x_min).replace(".", "p")
    x2 = "None" if x_max is None else str(x_max).replace(".", "p")
    return f"{window_name}_{x1}_{x2}"


# ============================================================
# LOAD EVERYTHING
# ============================================================
all_runs, meta_table = load_all_runs(data_folder, spectrum_block=spectrum_block_name)

print("All files metadata:")
display(meta_table)

if correction_mode == "no_background":
    sample_runs = [r for r in all_runs if r["row"]["base_name"] == sample_prefix]
    background_runs = []

    if len(sample_runs) == 0:
        raise ValueError(f"No sample files found with prefix '{sample_prefix}'")

    print(f"Found {len(sample_runs)} sample runs")
    print("No background files needed (mode = 'no_background')")

    runs_for_axis = sample_runs
else:
    sample_runs, background_runs = split_sample_background(
        all_runs,
        sample_prefix=sample_prefix,
        background_prefix=background_prefix,
    )

    print(f"Found {len(sample_runs)} sample runs")
    print(f"Found {len(background_runs)} background runs")

    runs_for_axis = sample_runs + background_runs

x_common, dx, decimals = build_global_common_axis(runs_for_axis)

print(f"\nGlobal common axis:")
print(f"dx = {dx} cm^-1")
print(f"decimals = {decimals}")
print(f"x range = [{x_common.min()}, {x_common.max()}]")
print(f"n points = {len(x_common)}")


# ============================================================
# BUILD CORRECTED RUNS
# ============================================================
if correction_mode == "no_background":
    run_rows = []
    for s in sample_runs:
        run_rows.append({
            "pol_deg": s["row"]["pol_deg"],
            "run_idx": s["row"]["run_idx"],
            "sample_file": s["file"].name,
        })

    run_table = pd.DataFrame(run_rows).sort_values(["pol_deg", "run_idx"]).reset_index(drop=True)
    print("\nSample runs used (no background mode):")
    display(run_table)

    corrected_runs = [
        process_one_sample_run_no_background(s, x_common)
        for s in sample_runs
    ]
else:
    pairs = pair_sample_background_runs(sample_runs, background_runs)

    pair_rows = []
    for p in pairs:
        pol_deg, run_idx = p["key"]
        pair_rows.append({
            "pol_deg": pol_deg,
            "run_idx": run_idx,
            "sample_file": p["sample"]["file"].name,
            "background_file": p["background"]["file"].name,
        })

    pair_table = pd.DataFrame(pair_rows)
    print("\nMatched sample/background pairs:")
    display(pair_table)

    corrected_runs = [
        correct_one_pair(p, x_common, mode=correction_mode)
        for p in pairs
    ]


# ============================================================
# AVERAGE CORRECTED RUNS BY POLARIZATION
# ============================================================
corrected_groups = group_corrected_runs_by_polarization(corrected_runs)

avg_by_pol = {}
for pol, group in corrected_groups.items():
    avg_by_pol[pol] = average_corrected_runs(group)
    print(f"Pol {pol:g} deg: averaged {avg_by_pol[pol]['n_runs']} corrected runs")


# ============================================================
# CREATE OUTPUT FOLDERS + SAVE AVERAGED SPECTRA
# ============================================================
if save_outputs:
    spectra_dir, figures_dir = ensure_output_dirs(output_folder)

    if save_avg_spectra_per_pol:
        save_avg_spectra(avg_by_pol, spectra_dir, correction_mode)


# ============================================================
# PLOT AVERAGED CORRECTED SPECTRA FOR ALL WINDOWS
# ============================================================
for window_name, x_min_win, x_max_win in x_windows:
    fig, ax = plt.subplots()

    for pol, d in avg_by_pol.items():
        x = d["x"]
        y = d["y_mean"]
        e = d["y_sem"]

        mask = get_window_mask(x, x_min_win, x_max_win)

        x_plot = x[mask]
        y_plot = y[mask]
        e_plot = e[mask]

        ax.plot(x_plot, y_plot, label=f"{pol:g}°")
        if np.any(np.isfinite(e_plot)):
            ax.fill_between(x_plot, y_plot - e_plot, y_plot + e_plot, alpha=0.2)

    ax.set_xlabel(r"Wavenumber (cm$^{-1}$)")
    if correction_mode == "ratio":
        ax.set_ylabel("Sample / Background")
        ax.set_title(f"Polarization-resolved corrected spectra (ratio) [{window_name}]")
    elif correction_mode == "subtract":
        ax.set_ylabel("Sample - Background")
        ax.set_title(f"Polarization-resolved corrected spectra (subtraction) [{window_name}]")
    else:
        ax.set_ylabel("Signal")
        ax.set_title(f"Polarization-resolved spectra (no background correction) [{window_name}]")

    ax.legend()
    plt.tight_layout()

    if save_outputs and save_main_average_plot:
        window_label = make_window_label(window_name, x_min_win, x_max_win)
        main_plot_file = figures_dir / f"all_polarizations_{correction_mode}_{window_label}.png"
        fig.savefig(main_plot_file, dpi=figure_dpi, bbox_inches="tight")
        print(f"Saved main average plot to: {main_plot_file}")

    plt.show()


# ============================================================
# OPTIONAL: PLOT ONE POLARIZATION, ALL INDIVIDUAL CORRECTED RUNS
# ============================================================
pol_to_show = sorted(avg_by_pol.keys())[0]

for window_name, x_min_win, x_max_win in x_windows:
    fig2, ax2 = plt.subplots()

    for r in corrected_groups[pol_to_show]:
        x = r["x"]
        y = r["y_corrected"]

        mask = get_window_mask(x, x_min_win, x_max_win)
        ax2.plot(x[mask], y[mask], alpha=0.7, label=f"Run {r['run_idx']}")

    d = avg_by_pol[pol_to_show]
    mask = get_window_mask(d["x"], x_min_win, x_max_win)

    ax2.plot(d["x"][mask], d["y_mean"][mask], lw=3, color="black", label="Average")

    ax2.set_xlabel(r"Wavenumber (cm$^{-1}$)")
    if correction_mode == "ratio":
        ax2.set_ylabel("Sample / Background")
        ax2.set_title(f"Corrected runs and average at {pol_to_show:g}° [{window_name}]")
    elif correction_mode == "subtract":
        ax2.set_ylabel("Sample - Background")
        ax2.set_title(f"Subtracted runs and average at {pol_to_show:g}° [{window_name}]")
    else:
        ax2.set_ylabel("Signal")
        ax2.set_title(f"Runs and average at {pol_to_show:g}° (no background) [{window_name}]")

    ax2.legend()
    plt.tight_layout()

    if save_outputs and save_single_pol_plot:
        window_label = make_window_label(window_name, x_min_win, x_max_win)
        single_plot_file = figures_dir / f"single_pol_{pol_to_label(pol_to_show)}deg_{correction_mode}_{window_label}.png"
        fig2.savefig(single_plot_file, dpi=figure_dpi, bbox_inches="tight")
        print(f"Saved single-polarization plot to: {single_plot_file}")

    plt.show()


# ============================================================
# EXTRA TOOL: COMPARE SELECTED POLARIZATION CURVES IN ONE PLOT
# ============================================================
def plot_compare_polarizations(
    avg_dict,
    pol_list=None,
    x_min=None,
    x_max=None,
    normalize=False,
    offset_step=0.0,
    show_error=True,
    title="Comparison of polarization curves",
    ylabel=None,
    save_plot=False,
    save_name=None,
):
    """
    avg_dict     : avg_by_pol dictionary from your pipeline
    pol_list     : list of polarizations to plot, e.g. [0, 20, 40, 60]
                   if None, plots all
    normalize    : if True, each curve is divided by its own max(abs(y))
    offset_step  : vertical offset added between curves for visibility
    show_error   : if True, plots SEM bands
    save_plot    : if True, saves figure into figures_dir
    save_name    : filename for saved figure
    """

    if pol_list is None:
        pol_list = sorted(avg_dict.keys())

    fig, ax = plt.subplots()

    for i, pol in enumerate(pol_list):
        if pol not in avg_dict:
            print(f"Warning: polarization {pol} not found, skipped.")
            continue

        d = avg_dict[pol]
        x = d["x"].copy()
        y = d["y_mean"].copy()
        e = d["y_sem"].copy()

        mask = get_window_mask(x, x_min, x_max)

        x = x[mask]
        y = y[mask]
        e = e[mask]

        if normalize:
            scale = np.nanmax(np.abs(y))
            if np.isfinite(scale) and scale > 0:
                y = y / scale
                e = e / scale

        if offset_step != 0:
            y = y + i * offset_step

        ax.plot(x, y, lw=2, label=f"{pol:g}°")

        if show_error and np.any(np.isfinite(e)):
            ax.fill_between(x, y - e, y + e, alpha=0.18)

    ax.set_xlabel(r"Wavenumber (cm$^{-1}$)")

    if ylabel is None:
        if correction_mode == "ratio":
            ylabel = "Sample / Background"
        elif correction_mode == "subtract":
            ylabel = "Sample - Background"
        else:
            ylabel = "Signal"

    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.legend()
    plt.tight_layout()

    if save_outputs and save_plot:
        if save_name is None:
            save_name = f"compare_selected_polarizations_{correction_mode}.png"
        compare_file = figures_dir / save_name
        fig.savefig(compare_file, dpi=figure_dpi, bbox_inches="tight")
        print(f"Saved comparison plot to: {compare_file}")

    plt.show()
```
